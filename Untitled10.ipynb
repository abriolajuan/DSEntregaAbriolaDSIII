En este trabajo final de Data Science, se realizara un análisis de sentimiento sobre una muestra de 100 tweets relacionados con Cristina Kirchner, específicamente en torno a la confirmación judicial de su condena.

Se aplican técnicas de procesamiento de lenguaje natural (NLP) utilizando Python, como limpieza de texto, lematización con spaCy, visualización de palabras frecuentes, vectorización con TF-IDF y un enfoque simbólico para el análisis de sentimiento.

```python
# Se importan librerías esenciales para manipular texto, y se carga el archivo con los tweets recolectados previamente.
import spacy
import pandas as pd
import re
from collections import Counter
import matplotlib.pyplot as plt

# Cargamos el modelo de español
nlp = spacy.load("es_core_news_sm")

# Carga del CSV con tweets
df = pd.read_csv("tweets_api_oficial.csv")

def limpiar_texto(texto):
    texto = re.sub(r'http\S+', '', texto)  # Quitar URLs
    texto = re.sub(r'@\w+', '', texto)     # Quitar menciones
    texto = re.sub(r'#', '', texto)        # Quitar hashtags
    texto = re.sub(r'\n', ' ', texto)      # Quitar saltos de línea
    texto = re.sub(r'[^A-Za-zÁÉÍÓÚáéíóúÑñÜü\s]', '', texto)  # Quitar caracteres no letras
    texto = texto.lower()                   # Pasar a minúsculas
    return texto.strip()

stopwords_personalizadas = {"cristina", "cfk", "kirchner", "macri", "milei", "cristin", "maximo", "fernandez"}

def preprocesar_texto(texto):
    doc = nlp(texto)
    tokens = []
    for token in doc:
        if (not token.is_stop
            and not token.is_punct
            and token.lemma_ != '-PRON-'
            and token.lemma_.lower() not in stopwords_personalizadas):
            tokens.append(token.lemma_)
    return " ".join(tokens)

df["texto_limpio"] = df["Contenido"].apply(limpiar_texto)
df["texto_procesado"] = df["texto_limpio"].apply(preprocesar_texto)

# Se analiza la frecuencia de palabras en los tweets preprocesados, para identificar los temas más mencionados.
all_tokens = [token for doc in df['texto_procesado'] for token in doc.split()]
counter = Counter(all_tokens)
most_common = counter.most_common(20)

palabras, frecuencias = zip(*most_common)

plt.figure(figsize=(12,6))
plt.bar(palabras, frecuencias)
plt.xticks(rotation=45)
plt.title('Palabras más frecuentes (sin nombres propios)')
plt.show()

print("Palabras más frecuentes filtradas:")
print(most_common)
```

```python
# Se genera una nube de palabras para visualizar los términos más recurrentes de forma atractiva e intuitiva
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = " ".join(df['texto_procesado'])

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import pandas as pd

#Se utiliza el método TF-IDF para identificar las palabras con mayor relevancia
# Vectorizador TF-IDF
vectorizer = TfidfVectorizer(max_features=1000)

# Ajustamos y transformamos los textos
tfidf_matrix = vectorizer.fit_transform(df["texto_procesado"])

# Obtenemos las palabras (features)
features = vectorizer.get_feature_names_out()

# Calculamos promedio de TF-IDF por palabra en todos los documentos
tfidf_means = tfidf_matrix.mean(axis=0).A1

df_tfidf = pd.DataFrame({
    'palabra': features,
    'tfidf_promedio': tfidf_means
})

df_tfidf = df_tfidf.sort_values(by='tfidf_promedio', ascending=False)

print(df_tfidf.head(20))
```

```python
#Se definen listas de palabras que reflejan posturas a favor o en contra de la condena, lo que permitirá clasificar los tweets de forma simbólica
# Posibles Palabras que indican apoyo al fallo (sentencia firme)
palabras_a_favor = ['condena', 'culpable', 'corrupta', 'corrupción', 'mafia', 'mafiosa', 'ladrona', 'robar', 'sentencia', 'justicia', 'fraude', 'delincuente', 'criminal']

# Posibles Palabras que indican rechazo al fallo (defensa de Cristina)
palabras_en_contra = ['apoyo', 'defender', 'inocente', 'honesta', 'mentira', 'persecución', 'odio', 'lawfare', 'arreglo', 'perseguida']
```

```python
def clasificar_sentimiento(texto):
    texto = texto.lower()
    favor = sum([1 for palabra in palabras_a_favor if palabra in texto])
    contra = sum([1 for palabra in palabras_en_contra if palabra in texto])

    if favor > contra:
        return 1  # A favor de la condena
    elif contra > favor:
        return -1  # En contra de la condena
    else:
        return 0  # Neutral o ambiguo
```

```python
df['sentimiento'] = df['texto_procesado'].apply(clasificar_sentimiento)
```

```python
import matplotlib.pyplot as plt

conteo = df['sentimiento'].value_counts().sort_index()
labels = ['En contra del fallo (-1)', 'Neutral (0)', 'A favor del fallo (+1)']
plt.bar(labels, conteo)
plt.title('Análisis de Sentimiento sobre Cristina y la condena')
plt.ylabel('Cantidad de Tweets')
plt.xticks(rotation=15)
plt.show()
```

```python
porcentajes = df['sentimiento'].value_counts(normalize=True) * 100
print(porcentajes.sort_index())
```

A partir del relevamiento de 100 tweets recientes relacionados con Cristina Kirchner y la confirmación de su condena, se realizó un análisis de sentimiento simbólico, utilizando listas personalizadas de palabras con carga positiva o negativa respecto del fallo judicial.

Los resultados obtenidos muestran que:

El 27% de los tweets expresan una postura a favor de la condena, con términos como “condena”, “culpable”, “corrupción”.

Un 3% manifiesta una postura en contra del fallo, con palabras como “inocente”, “persecución”, “lawfare”.

El 70% restante contiene expresiones neutrales o ambiguas, sin una orientación clara.

Esta distribución sugiere que, dentro de esta muestra limitada, predomina la neutralidad, aunque hay una mayor proporción de mensajes críticos hacia la exmandataria que defensivos.
